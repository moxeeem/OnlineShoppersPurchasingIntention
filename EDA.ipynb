## Target Variable Analysis

The dataset has a target variable called `Revenue`. It is a categorical variable that indicates whether a visitor made a purchase or not.

The variable has a Boolean data type, which means it can take two values: `True` or `False`. If a visitor made a purchase, the value of `Revenue` will be `True`, otherwise, it will be `False`.

As the target variable, `Revenue` will be used in the analysis to predict whether a visitor is likely to make a purchase or not.

Let's look at the distribution of the variable

````python
sns.countplot(x='Revenue',data=df)
````

As we can see, this dataset is imbalanced

````python
df['Revenue'].value_counts(normalize=True) * 100
````

We see an offset towards the False class.

*   With unbalanced classes, models may lean towards a class that has a larger number of objects. As a result, the model may show poor performance on a smaller class.
*   If the classes are not balanced, then the model can easily retrain on training data, showing high accuracy on training data and poor performance on new data.
*   An unbalanced target variable can lead to a decrease in the overall accuracy of the model.

````python
sns.histplot(data = df['Revenue'], kde = True, bins = 2)
````

If we look at the distribution graph, it will resemble the Bernoulli distribution.

image.png

However, it should be borne in mind that the Bernoulli distribution assumes independence between individual observations, which may not be fulfilled in the case of the target variable

## Pearsons Chi-Square Test

Let's try to apply Chi-Square test for some features.

````python
from scipy.stats import chi2_contingency
````

### Month and Revenue

````python
cross_tab = pd.crosstab(df['Revenue'], df['Month'])

chi2, p, dof, expected = chi2_contingency(cross_tab)

print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
````

The p-value equal to 1.4522308860506912e-74 is very small, which indicates a statistically significant relationship between the variables, since it is less than the usually selected significance level of 0.05.

### Region and Revenue

````python
cross_tab = pd.crosstab(df['Revenue'], df['Region'])

chi2, p, dof, expected = chi2_contingency(cross_tab)

print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
````

Based on the results of the Pearson chi-squared test, it can be concluded that there is no statistically significant relationship between the variables Revenue and Region at the significance level of 0.05, since p-value is 0.312, which is greater than the selected significance level.

### VisitorType and Revenue

````python
cross_tab = pd.crosstab(df['Revenue'], df['VisitorType'])

chi2, p, dof, expected = chi2_contingency(cross_tab)

print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)
````

The p-value equal to 6.934250931509824e-29 is very small, which indicates a statistically significant relationship between the variables, since it is less than the usually selected significance level of 0.05.

### Introduction

Let's increase the size of the dataset by a simple duplication by 101 times and compare the speed of the Pandas and Polars libraries.

````python
df_big = df.copy()

for i in range(100):
  df_big = pd.concat([df_big, df])

df_big.to_csv("purchasing_big.csv", index=False)
````

````python
df_big.shape, df.shape
````

We have increased the dataset 101 times

### Loading Data

````python
# Polars
%%time

df_pl = pl.read_csv("purchasing_big.csv")
````

````python
# Pandas
%%time

df_pd = pd.read_csv("purchasing_big.csv")
````

As we can see, the data loading speed differs by almost 2 times. Polars loads faster.

### Data Filtering

````python
# Pandas
%%time

df_pd.query('OperatingSystems == 3')
````

````python
# Polars
%%time

df_pl.filter(pl.col('OperatingSystems') == 3)
````

As we can see, the data filtering speed differs by almost 1,5 times. Pandas loads faster

### Data Aggregation

````python
# Pandas

%%time

df_pd.groupby('Administrative_Duration').agg({'Administrative' : 'mean', 'ProductRelated_Duration' : 'max'})
````

````python
# Polars

%%time

q = (
    df_pl
    .lazy()
    .groupby(by='Administrative_Duration')
    .agg(
        [
            pl.col('Administrative').mean(),
            pl.col('ProductRelated_Duration').max()
        ]
    )
    )
````

````python
%%time

q.collect()
````

As we can see, the data filtering speed differs. Polars works a little faster
